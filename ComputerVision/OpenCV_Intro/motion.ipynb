```python
import cv2

cap = cv2.VideoCapture(0)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

print("Motion Detection Active. Press 'q' to quit.")

while cap.isOpened():
    #Find the absolute mathematical difference between the two frames
    diff = cv2.absdiff(frame1, frame2)
    
    #Convert to grayscale (makes the math easier)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    
    #Blur it slightly to ignore tiny camera noises
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    
    #Thresholding: If a pixel difference is greater than 20, make it pure white (255)
    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)
    
    #Show the resulting "Motion Mask" on the screen
    cv2.imshow("Motion Detector (White = Movement)", thresh)
    cv2.imshow("Original Camera Feed", frame1)
    
    #Shift the frames forward in time
    frame1 = frame2
    ret, frame2 = cap.read()
    
    # Press 'q' to break the loop and close the window
    if cv2.waitKey(40) == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```
