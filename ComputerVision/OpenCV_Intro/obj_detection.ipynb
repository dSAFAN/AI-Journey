```python
import cv2

# 1. Load the pre-trained Haar Cascade XML file (The "Rulebook" for faces)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# 2. Open the webcam
cap = cv2.VideoCapture(0)

print("Face Detection Active. Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    
    # Convert to grayscale (Cascades only work on black-and-white contrasts)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 3.Scan the image for face patterns
    # scaleFactor: Compensates for faces appearing smaller when further away
    # minNeighbors: Requires multiple overlapping detections to confirm it's a real face
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    # 4. Draw a green rectangle around any face the computer finds
    # faces returns a list of coordinates: (x, y, width, height)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
    # Show the result
    cv2.imshow('Face Tracker', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
