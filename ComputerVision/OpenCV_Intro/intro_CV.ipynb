```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the image using OpenCV
# OpenCV reads color images in BGR (Blue, Green, Red) format by default, not RGB.
img_bgr = cv2.imread('images/braille.jpeg') 

# 2. Convert it to standard RGB so Matplotlib can draw it correctly
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 3. Print the raw mathematical shape of the image
print(f"Image Shape: {img_rgb.shape}") 

# 4. Display it
plt.imshow(img_rgb)
#plt.imshow(img_bgr)
plt.axis('off') # Hides the axis numbers
plt.show()
```

```python
# 1. Convert the image to Grayscale (simplifies the math from 3D to 2D)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# 2. Define our custom Kernel (A 3x3 matrix)
# This specific pattern of numbers calculates differences between neighboring pixels.
# If pixels are similar (flat color), it outputs 0 (black). 
# If pixels are very different (an edge), it outputs a high number (white).
edge_kernel = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# 3. Apply the sliding window (filter) to the image
edges = cv2.filter2D(img_gray, -1, edge_kernel)

# 4. Display the result
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```
