```python
import collections
import collections.abc
collections.Mapping = collections.abc.Mapping
from experta import *
```

```python
# ==========================================
# 1. THE ONTOLOGY (Classes & Data Properties)
# ==========================================

class PersonState(Fact):
    """Tracks physical/mental state. Properties: energy_level, time_of_day"""
    pass

class CurrentActivity(Fact):
    """Tracks the current task. Properties: activity_category, specific_task, duration_mins"""
    pass
```

```python
# ==========================================
# 2. THE KNOWLEDGE BASE & INFERENCE ENGINE
# ==========================================

class MyDayExpertSystem(KnowledgeEngine):
    
    # --- MORNING ROUTINE ---
    @Rule(AND(
        PersonState(time_of_day="Morning"),
        PersonState(energy_level="High")
    ))
    def morning_commute_rule(self):
        print("💡 INFERENCE: Morning time with high energy.")
        print("🎯 ACTION: Initiate Commute. Time to head to the onsite internship!\n")

    # --- WORK/INTERNSHIP LOGIC ---
    @Rule(AND(
        CurrentActivity(activity_category="Work"),
        CurrentActivity(duration_mins=P(lambda x: x > 240)) 
    ))
    def deep_work_fatigue_rule(self):
        print("💡 INFERENCE: You have been at the onsite internship for over 4 hours.")
        print("🎯 ACTION: Declare 'PersonState' as Fatigued. Recommend a lunch break or Walk.\n")
        # The engine can declare new facts based on rules!
        self.declare(PersonState(energy_level="Low"))

    # --- STUDY LOGIC (College & IT) ---
    @Rule(AND(
        CurrentActivity(activity_category="Study"),
        CurrentActivity(specific_task="PostgreSQL/Python")
    ))
    def it_study_rule(self):
        print("💡 INFERENCE: IT Study session detected.")
        print("🎯 ACTION: Open GitHub. Remember to commit changes and check repository structures.\n")

    @Rule(AND(
        CurrentActivity(activity_category="Study"),
        CurrentActivity(specific_task="BCA Coursework")
    ))
    def college_study_rule(self):
        print("💡 INFERENCE: College coursework session detected.")
        print("🎯 ACTION: Focus on Cluster University syllabus. Target that 7.0+ CGPA!\n")

    # --- LEISURE & RECOVERY LOGIC ---
    @Rule(PersonState(energy_level="Low"))
    def recovery_rule(self):
        print("💡 INFERENCE: Energy levels have dropped to Low.")
        print("🎯 ACTION: Transition to Rest. Fire up the PPSSPP emulator for some WWE 2K or PES 2016.\n")

print("Ontology mapped and Expert System initialized!")
```

```python
# Instantiate the engine
my_day_ai = MyDayExpertSystem()

print("--- SCENARIO 1: The Morning Start ---")
my_day_ai.reset() # Clears memory
my_day_ai.declare(PersonState(time_of_day="Morning", energy_level="High"))
my_day_ai.run()

```

```python
print("--- SCENARIO 2: The Long Shift ---")
my_day_ai.reset()
# We inject a 5-hour work shift (300 minutes)
my_day_ai.declare(CurrentActivity(activity_category="Work", duration_mins=300))
my_day_ai.run()
```

```python

print("--- SCENARIO 3: Evening Grind ---")
my_day_ai.reset()
my_day_ai.declare(CurrentActivity(activity_category="Study", specific_task="PostgreSQL/Python"))
my_day_ai.run()
```
