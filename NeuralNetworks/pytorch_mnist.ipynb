```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader


# 1. GET THE DATA
# We convert the images into "Tensors" (PyTorch's GPU-ready version of NumPy arrays)
transform = transforms.Compose([transforms.ToTensor()])

train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# DataLoaders automatically shuffle the images and feed them to the AI in "batches" of 64
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# 2. BUILD THE ARCHITECTURE (The LEGO Blocks)
ai_model = nn.Sequential(
    nn.Flatten(),           # Flattens the 28x28 2D image into a 1D line of 784 pixels
    nn.Linear(784, 128),    # Hidden Layer: 784 inputs connect to 128 brain cells
    nn.ReLU(),              # Activation Function: Adds curves to our math
    nn.Linear(128, 10)      # Output Layer: 128 cells vote on 10 possible digits (0-9)
)

# 3. DEFINE THE RULES
criterion = nn.CrossEntropyLoss() # Our grader (Loss Function)
optimizer = optim.Adam(ai_model.parameters(), lr=0.001) # The weight updater (Adam is faster than SGD)

# 4. THE TRAINING LOOP
epochs = 5 # We will read the entire dataset 5 times

print("\n--- Starting Training ---")
for epoch in range(epochs):
    running_loss = 0.0
    
    # Loop through the data in batches of 64 images at a time
    for images, labels in train_loader:
        
        # Step A: Clear old gradients
        optimizer.zero_grad()
        
        # Step B: Forward Pass (Make a guess)
        outputs = ai_model(images)
        
        # Step C: Calculate the Error
        loss = criterion(outputs, labels)
        
        # Step D: Backpropagation (The Autograd Magic!)
        loss.backward()
        
        # Step E: Update the Weights
        optimizer.step()
        
        running_loss += loss.item()
        
    print(f"Epoch {epoch+1}/{epochs} Completed | Average Error (Loss): {running_loss/len(train_loader):.4f}")

print("Training Complete!")
```

```python
# 5. EVALUATE THE MODEL
correct = 0
total = 0

# Turn off the Autograd engine because we are just testing, not learning
with torch.no_grad():
    for images, labels in test_loader:
        outputs = ai_model(images)
        
        # Find the neuron (0-9) that shouted the loudest
        _, predicted = torch.max(outputs.data, 1)
        
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Final AI Accuracy on Unseen Data: {accuracy:.2f}%")
```
