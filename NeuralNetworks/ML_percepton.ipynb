```python
import numpy as np
```

```python

class MultiLayerPerceptron:
    def __init__(self, input_nodes=2, hidden_nodes=4, output_nodes=1, learning_rate=0.5):
        self.lr = learning_rate
        
        # 1. Initialize Weights and Biases for the Hidden Layer
        # Matrix shape: (2 inputs x 4 hidden neurons)
        self.weights_hidden = np.random.uniform(-1, 1, (input_nodes, hidden_nodes))
        self.bias_hidden = np.random.uniform(-1, 1, (1, hidden_nodes))
        
        # 2. Initialize Weights and Biases for the Output Layer
        # Matrix shape: (4 hidden neurons x 1 output neuron)
        self.weights_output = np.random.uniform(-1, 1, (hidden_nodes, output_nodes))
        self.bias_output = np.random.uniform(-1, 1, (1, output_nodes))
    
    # The Non-Linear Activation Function
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    # The derivative of Sigmoid (Used for Backpropagation to calculate gradients)
    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def train(self, X, y, epochs=10000):
        print("Starting Training...")
        
        for epoch in range(epochs):
            # ==========================================
            # STEP 1: THE FORWARD PASS (Making a guess)
            # ==========================================
            
            # Input to Hidden Layer
            hidden_layer_input = np.dot(X, self.weights_hidden) + self.bias_hidden
            hidden_layer_activation = self.sigmoid(hidden_layer_input)
            
            # Hidden to Output Layer
            output_layer_input = np.dot(hidden_layer_activation, self.weights_output) + self.bias_output
            predicted_output = self.sigmoid(output_layer_input)
            
            # ==========================================
            # STEP 2: BACKPROPAGATION (The Blame Game)
            # ==========================================
            
            # How wrong was the final output?
            error = y - predicted_output
            
            # Output Layer Blame: Error * derivative of activation
            d_predicted_output = error * self.sigmoid_derivative(predicted_output)
            
            # Hidden Layer Blame: How much did the hidden weights contribute to the output error?
            error_hidden_layer = d_predicted_output.dot(self.weights_output.T)
            d_hidden_layer = error_hidden_layer * self.sigmoid_derivative(hidden_layer_activation)
            
            # ==========================================
            # STEP 3: UPDATE WEIGHTS & BIASES
            # ==========================================
            
            # Update Output Layer weights
            self.weights_output += hidden_layer_activation.T.dot(d_predicted_output) * self.lr
            self.bias_output += np.sum(d_predicted_output, axis=0, keepdims=True) * self.lr
            
            # Update Hidden Layer weights
            self.weights_hidden += X.T.dot(d_hidden_layer) * self.lr
            self.bias_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * self.lr
            
            # Print progress every 2000 epochs
            if epoch % 2000 == 0:
                loss = np.mean(np.square(error))
                print(f"Epoch {epoch} - Loss (Error Rate): {loss:.4f}")

    def predict(self, X):
        # Run a quick forward pass to get a prediction
        hidden_layer_activation = self.sigmoid(np.dot(X, self.weights_hidden) + self.bias_hidden)
        predicted_output = self.sigmoid(np.dot(hidden_layer_activation, self.weights_output) + self.bias_output)
        return np.round(predicted_output) # Round to 0 or 1

print("MLP Architecture Loaded!")
```

```python
# The XOR Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# The Target Outputs for XOR
y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Instantiate and Train!
mlp = MultiLayerPerceptron(learning_rate=0.5)
mlp.train(X, y, epochs=10000)

print("\n--- Testing the Trained MLP on XOR ---")
print(f"Input [0, 0] -> Predicts: {mlp.predict(np.array([[0, 0]]))[0][0]}")
print(f"Input [0, 1] -> Predicts: {mlp.predict(np.array([[0, 1]]))[0][0]}")
print(f"Input [1, 0] -> Predicts: {mlp.predict(np.array([[1, 0]]))[0][0]}")
print(f"Input [1, 1] -> Predicts: {mlp.predict(np.array([[1, 1]]))[0][0]}")
```
