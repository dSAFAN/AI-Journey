```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SingleLayerPerceptron:
    def __init__(self, learning_rate = 0.1, epochs = 10):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        self.errors_per_epoch = []
    
    def activation_function(self,z):
        return np.where(z > 0, 1 , 0)
    
    def fit(self, X , y):
        
        """
        x = inputs (training data)
        y = correct answers (target)
        """

        self.weights = np.random.uniform(-0.5,0.5,X.shape[1])
        self.bias = 0.0

        print(f"Starting Training! Initial Random Weights: {self.weights}, Bias: {self.bias}\n")

        for epoch in range(self.epochs):
            errors = 0

            for inputs, target in zip(X,y):
        
                # Multiply inputs by weights and add bias 
                z = np.dot(inputs, self.weights) + self.bias
                prediction = self.activation_function(z)
                
                # STEP 3: Calculate the Error
                error = target - prediction

                # STEP 4: Update the Weights and Bias
                # Formula: W_new = W_old + (learning_rate * error * input)
                if error != 0:
                    self.weights += self.learning_rate * error * inputs
                    self.bias += self.learning_rate * error
                    errors += 1 # Logs that we made a mistake
            
            self.errors_per_epoch.append(errors)
            print(f"Epoch {epoch + 1}/{self.epochs} - Mistakes made: {errors}")
            
            # If we make 0 mistakes, the Perceptron has perfectly learned the rules!
            if errors == 0:
                print(f"\nTraining Complete! The Perceptron figured it out.")
                print(f"Final Trained Weights: {self.weights}, Final Bias: {self.bias}")
                break

    def predict(self, X):
        # Once trained, we use this to ask the network for answers
        z = np.dot(X, self.weights) + self.bias
        return self.activation_function(z)

print("Perceptron Architecture Loaded!")
```

```python
# 1. Define the Training Data (The AND Gate)

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# y is the correct answers we want the machine to learn
y = np.array([0, 0, 0, 1])

# 2. Instantiate our AI
# We give it a learning rate of 0.1, and give it 20 tries (epochs) to figure it out
ai_model_and = SingleLayerPerceptron(learning_rate=0.1, epochs=20)

# 3. Train it!
ai_model_and.fit(X, y)
```

```python

# 4. Let's graph the learning progress to visually prove it worked
plt.plot(range(1, len(ai_model_and.errors_per_epoch) + 1), ai_model_and.errors_per_epoch, marker='o')
plt.xlabel('Epochs (Tries)')
plt.ylabel('Number of Mistakes')
plt.title('Perceptron Learning Progress')
plt.grid(True)
plt.show()

# 5. Testing the trained model
print("\n--- Testing the Brain ---")
print(f"Input [1, 1] -> AI Predicts: {ai_model_and.predict(np.array([1, 1]))}")
print(f"Input [0, 1] -> AI Predicts: {ai_model_and.predict(np.array([0, 1]))}")
```
